```python
import numpy as np
import matplotlib.pyplot as plt

def intersection_nomograph_lin(v3_values, v1_range):
    fig, ax = plt.subplots(figsize=(8, 8))

    v1 = np.linspace(v1_range[0], v1_range[1], 1000)  # Define range of v1
    for v3 in v3_values:
        v2 = v3 / v1  # Calculate v2 using v3 - v2*v1 = 0
        ax.plot(v1, v2, color='black', linewidth=1)  # Plot the curve for each v3 value

    ax.set_xlabel('v1')
    ax.set_ylabel('v2')
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True, which='both', linestyle='-', linewidth=1, color='black')  # Dense grid with small intervals

    ax.set_xticks(np.arange(0, 10, 0.5))
    ax.set_yticks(np.arange(0, 10, 0.5))

    ax.set_xlim(0, 10)  # Limiting x-axis from 0 to 2
    ax.set_ylim(0, 10)  # Limiting y-axis from 0 to 10
    ax.grid(True, linestyle='-', linewidth=0.4, alpha=1)  # Thin and dense grid
    plt.savefig('intersection_nomograph.svg', format='svg')  # Save plot as SVG
    plt.show()

# Example usage
v3_values = range(5, 100, 5)  # Varying v3 values
v1_range = (0.1, 10)  # Range of v1
intersection_nomograph_lin(v3_values, v1_range)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def intersection_nomograph(v3_values, v1_range):
    fig, ax = plt.subplots(figsize=(8, 8))

    v1 = np.linspace(v1_range[0], v1_range[1], 1000)  # Define range of v1
    for v3 in v3_values:
        v2 = v3 / v1  # Calculate v2 using v3 - v2*v1 = 0
        ax.plot(v1, v2, color='black', linewidth=1)  # Plot the curve for each v3 value

    ax.set_xlabel('v1')
    ax.set_ylabel('v2')
    ax.set_aspect('equal', adjustable='box')
    ax.grid(True, which='both', linestyle='-', linewidth=1, color='black')  # Dense grid with small intervals

    ax.set_xscale('log')  # Logarithmically spaced x ticks
    ax.set_yscale('log')  # Logarithmically spaced y ticks

    ax.set_xlim(1, 10)  # Limiting x-axis from 1 to 10
    ax.set_ylim(1, 10)  # Limiting y-axis from 0.1 to 10

    ax.grid(True, linestyle='-', linewidth=0.4, alpha=1)  # Thin and dense grid
    plt.savefig('intersection_nomograph.svg', format='svg')  # Save plot as SVG
    plt.show()

# Example usage
v3_values = range(1 , 100, 1)  # Varying v3 values
v1_range = (0.1, 10)  # Range of v1
intersection_nomograph(v3_values, v1_range)
```
